```python
import random


class Creature:
    def __init__(self, name, max_hp = 10 , attack=1, defense=5, speed=5):
        self.name = name
        self.max_hp = max_hp
        self.hp = max_hp
        self.abilities = {"attack": attack, "defense": defense, "speed": speed}

    def check_life(self):
        if self.hp <= 0:
            print(f"{self.name} fainted...")
            self.hp = 0
            return 0
        else:
            return self.hp

    def attack(self, target):
        roll = random.randint(1, 20)
        if roll < target.abilities["defense"] + target.abilities["speed"]:
            print(f"Attack missed...")
            return False
        else:
            damage = self.abilities["attack"] + random.randint(1, 4)
            target.hp -= damage
            print(f"Attack hits for {damage} damage!")
            return True

    def turn(self, target):
        if self.attack(target):
            if target.check_life() == 0:
                return True
        return False

gollum = Creature("Gollum", 10)
bilbo = Creature("Bilbo", 10)

round_num = 1
while round_num <= 20:
    print(f"Round {round_num}:")
    print(f"{gollum.name} attacks {bilbo.name}.")
    if gollum.turn(bilbo):
        print(f"{bilbo.name} has been defeated!")
        break
    print(f"{bilbo.name} attacks {gollum.name}.")
    if bilbo.turn(gollum):
        print(f"{gollum.name} has been defeated!")
        break
    round_num += 1
```
