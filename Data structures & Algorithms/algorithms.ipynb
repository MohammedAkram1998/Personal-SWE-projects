```python
def fact_iterative(y):
    
    x = 1 
    
    
    for i in range(1 , y+1):
        x = x*i 
    return x 

def fact_tailrecursive(y, a= 1):
    if y==1 or  y==0 :
        return a 
    
    
    
    return fact_tailrecursive(y-1   , (a*y))
                              
                              
def fact_nontailrecursive(y):
    if y==1 or y==0:
        return 1 
                              
    return (fact_nontailrecursive(y-1))*(y)

print(fact_iterative(10))
print(fact_tailrecursive(10))
print(fact_nontailrecursive(10))
    
```

```python
import time 


#make 3 empty lists to start off with
runtime_nontailrecursive = []
runtime_tailrecursive = []
runtime_iterative = []


# set the inputs in a variable valueus and the number of repeats to 1 million 
values = list(range(1,20))
repeats =   1000000

# make 3 empty list for the run times. we will divide by repetitions later 
for value in values:
    runtime1_nontailrecursive = []
    runtime1_tailrecursive = []
    runtime1_iterative = []
    
    
    
    
    
    for i in range(repeats):
        # we are going to measure the time at s 
        # then call the function and measure it at time z 
        # this will give us the runtime 
        s = time.time()
        fact_iterative(value)
        z = time.time()
        runtime1_iterative.append(z-s)
        
        
        # we are going to measure the time at s 
        # then call the function and measure it at time z 
        # this will give us the runtime 
        s = time.time()
        fact_tailrecursive(value)
        z = time.time()
        runtime1_tailrecursive.append(z-s)
        
        # we are going to measure the time at s 
        # then call the function and measure it at time z 
        # this will give us the runtime 
        s = time.time()
        fact_nontailrecursive(value)
        z = time.time()
        runtime1_nontailrecursive.append(z-s)
        
        
# the next thing we are going to do is to get the runtimes and divide by the number 
# of repetitons so we can get the average runtime 

# first change the nontail recursive time to average time 

    runtime_nontailrecursive.append(sum(runtime1_nontailrecursive)  /  repeats )

# then change the tail recirsive time to average time 
    runtime_tailrecursive.append(sum(runtime1_tailrecursive)  /  repeats )

# finally change the iterative time to the average time 
    runtime_iterative.append(sum(runtime1_iterative)  /  repeats )

    
```

```python
import matplotlib.pyplot as matt 
# start off by creating the 3 lines for the 3 algorithsm 
matt.plot(values,   runtime_iterative ,  label = "iterative ")
matt.plot(values,   runtime_tailrecursive ,  label = "tail recursive ")
matt.plot(values,   runtime_nontailrecursive ,  label = "non tail recursive ")
# set the x axis and y axis labels 
matt.ylabel("running time in microseconds")
matt.xlabel("fact number")
# show the graph 
matt.legend()
matt.show 
```

Analysis 

the graph above compares the input values for the factorial function and how long it takes for an algorithm to take the input and calculate the factorial of that number(averaged over many repetitions) 

for example if n=10 and the avergae runtime is 1 microsecond then that means it took on average that algorithm 1 microsecond to calculate the 10!.

The graph shows how the average runtimes for some of the alorgoithms are lower than others for values of input in range 1 to 20. Lower runtimes means higher algorithmic efficiency here.
The most efficient algotihm is the iterative followed by non tail recursive followed by tail recursive. This result is somewhat unusual because I would have expected the tail recursive algorithm to be more efficient as it can be used by an interpreter to reuse the existing stack frame[1]. However this can vary between different python interpreters and its possible the interpreter being used here doesnt optimise tail recursion which explains why the iterative algorithm is more efficient. I noticed that when I used large numbers of repeats the grpah was less spiky than with smaller number of repeats. This is because average runtimes will correct for any random or noise based erors and make your mesaurements more accurate.

```python
def linearsearch(a, b):
    
    for i, x in enumerate(a):
        
        if x == b:
            return i 
    return -1 



def binarysearch(a,b):
    x, y = 0 , len(a)-1 
    
    
    while x <= y :
        middle = ((x+y)//2)
        
        if a[middle]== b :
            return middle
        
        elif a[middle]< b:
            x = middle +1 
        
        
        else:
            y = middle -1 
    return -1
```

```python
import time 


lineartime = []
binarytime = []


x = runtime_iterative 

sortedx = sorted(x)

repeats = 1000000


for i in x:
    lineartime1 = []
    binarytime1 = []
    
    for u in range(repeats):
        # start with finding the time for linear search
        t1 = time.time()
        linearsearch (x,i)
        t2 = time.time()
        lineartime1.append(t2-t1)
        
        # then find the time for binary search.
        t1 = time.time()
        binarysearch (sortedx,i)
        t2 = time.time()
        binarytime1.append(t2-t1)
    
    # divide by number of repeats to get the average time 
    lineartime.append(sum(lineartime1)  /  repeats )
    binarytime.append(sum(binarytime1)  /  repeats )
        
matt.plot(sortedx, lineartime ,  label = "linear seach ")
matt.plot(sortedx, binarytime ,  label = "binary  search ")

# set the x and y axis labels 
matt.ylabel("time in seconds (x 10^-7)")
matt.xlabel("n") 
# plot the graph 
matt.legend()    
matt.show()
        
        
```
